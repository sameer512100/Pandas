<h1>Loading data into pandas</h1>

```python
import pandas as pd
poke = pd.read_csv("pokemon.csv")
# poke.head(5)
```

<h1>Reading Data in Pandas</h1>

<h3>Read Headers</h3>

```python
print(poke.columns)
```

<h3>Read Each Column</h3>

```python
print(poke['Name'])
```

```python
# PARTICULAR COLS
print(poke['Name'][0:5])
```

```python
print(poke[["Name","Type 1","HP"]])
```

<h3>Read Each Row</h3>

```python
poke.head(5)
```

```python
# ILOC = Integer Location
print(poke.iloc[1])
```

```python
print(poke.iloc[1:5])
```

```python
print(poke.iloc[1,1]) ## 1ST ROW 1ST ELEM
```

```python
# LOOP
for index,row in poke.iterrows():
    print(index,row['Type 1'])
```

```python
poke.loc[poke['Type 1']=='Fire']
```

<h1>SORTING/DESCRIBING DATA</h1>

```python
poke.describe()
```

```python
poke.sort_values('Name') # ALPHABETICAL SORT
```

```python
poke.sort_values('Name',ascending=False) ## REVERSE SORT
```

```python
## SORT FOR MULTIPLE PARAMETERS
poke.sort_values(['Type 1','HP'],ascending=False)
```

```python
poke.sort_values(['Type 1','HP'],ascending=[1,0]) ## 1 = ascending , 0 = descending
```
